# Doughnut of Social and Planetary Boundaries

In a [recent paper in Nature](https://www.nature.com/articles/s41586-025-09385-1), Andrew L. Fanning and Kate Raworth examined the progress, or lack thereof, of meeting the needs of all people within the means of the living planet. To do that, they used a set of 35 indicators that monitor trends in social deprivation and ecological overshoot over the 2000–2022 period. They found that, although global gross domestic product (GDP) has more than doubled, there has been only a modest achievement in reducing human deprivation; at the same time, the increase in ecological overshoot does not bode well for Earth-system stability.

In this assignent, you will work through some of their results yoursefl. You will use their data, available at <https://doi.org/10.5281/zenodo.15688961>.
